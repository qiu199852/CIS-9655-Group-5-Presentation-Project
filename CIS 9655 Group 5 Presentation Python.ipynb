```python
data = pd.read_csv('Covid-19.csv', index_col=0, parse_dates=True)
case = data['CASE_COUNT']
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(1, 1, 1)
case.plot(ax=ax, style='k-')
plt.title('March 2020 - March 2022 NYC COVID-19 Cases',fontweight="bold")
plt.xlabel("Month of the Year")
plt.ylabel("Number of COVID-19 Cases")
plt.annotate(xy=["Dec 2021", -1000], text = 'Christmas/New Year', color = 'g')
plt.annotate(xy=["Dec 2020", -1000], text = 'Christmas/New Year',color = 'g')
plt.annotate(xy=["Jul 2021", -1000], text = 'Back to School/In-person',color = 'g')
plt.annotate(xy=["Apr 2020", -1000], text = 'COVID-19 Start',color = 'g')

plt.annotate(xy=["Dec 2020", 15000], text = 'Increase cases',color = 'r',fontsize='14')
plt.annotate(xy=["Dec 2020", 13000], text = 'around holiday',color = 'r',fontsize='14')

plt.annotate(xy=["Oct 2021", 40000], text = 'Highest peak',color = 'r',fontsize='14')
plt.annotate(xy=["Oct 2021", 38000], text = 'around holiday-',color = 'r',fontsize='14')
plt.annotate(xy=["Oct 2021", 36000], text = 'Omicorn',color = 'r',fontsize='14')

plt.show()
```

```python
case = data['HOSPITALIZED_COUNT']
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(1, 1, 1)
case.plot(ax=ax, style='b')
plt.title('March 2020 - March 2022 Hospitalized Cases vs. Death Cases',fontweight="bold")
plt.xlabel("Month of the Year")
plt.ylabel("Number of COVID-19 Cases")

case2 = data['DEATH_COUNT']
case2.plot(ax=ax, style='r')

plt.annotate(xy=["Dec 2021", -45], text = 'Christmas/New Year', color = 'g')
plt.annotate(xy=["Dec 2020", -45], text = 'Christmas/New Year',color = 'g')
plt.annotate(xy=["Jul 2021", -45], text = 'Back to School/In-person',color = 'g')
plt.annotate(xy=["Apr 2020", -45], text = 'COVID-19 Start',color = 'g')
plt.legend(['Hospitalized','Death'])
plt.xlabel("Month of the Year")
plt.ylabel("Number of COVID-19 Cases")
            
plt.show()
```
